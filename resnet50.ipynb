# Simple ResNet50 V2 example

This example is tested on TensorFlow 2.7.0

```python
image_see = True  # If you want to see image, set it True
data_see = False  # If you want to see data, set it True

if image_see:
    %matplotlib inline
    import cv2
    from matplotlib import pyplot as plt
```

```python
# Load necessary functions

import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.resnet_v2 import ResNet50V2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet_v2 import preprocess_input, decode_predictions
```

```python
#Load model

model = ResNet50V2(weights='imagenet')
```

```python
#Load image and preprocess
img_path = 'dog.jpg'

img = image.load_img(img_path, target_size = (224, 224))
x = image.img_to_array(img)
print(x.shape)

if image_see: plt.imshow(img)
```

```python
x = np.expand_dims(x, axis = 0)
print(x.shape)
if (data_see == True): print(x)
```

Preprocess image : subtract and divide using RGB-mean, scale

```python
x = preprocess_input(x)
if (data_see == True): print(x)
```

```python
%%time
# %%time is to measure elapsed time
preds = model.predict(x)
```

```python
print("Predicted: ", decode_predictions(preds, top = 3)[0])
```
